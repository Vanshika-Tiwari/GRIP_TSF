<h1 style="background-color:#ABFF9B;font-size:20px;color:#00033E;font-weight : bold"> <B>Problem Statement:</B> Creating a <B>Decision Tree Classifier</B> of Iris Dataset and visualizing it graphically.</h1>


```python
#Importing required libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
#Importing the Dataset

data = pd.read_csv("Iris.csv")
```

```python
#Printing the first 5 rows
data.head()
```

```python
#Checking the number of rows and columns of the data
iris = pd.DataFrame(data)
iris.shape
```

```python
iris.describe()
```

```python
#Checking null values (if any)
iris.isnull().sum()
```

```python
x = iris.drop(["Species", "Id"], axis= 1)
y = iris.Species
print(x)
print(y)
```

```python
iris.Species.unique()
```

```python
#Training and fitting the model

dt = DecisionTreeClassifier(max_depth=5)
dt.fit(x,y)
```

```python
#Visualizing the Decision Tree

yn = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
xn = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi=300)

tree.plot_tree(dt,
           feature_names = xn, 
           class_names = yn,
           filled = True);

#Save the image if required
#fig.savefig('imagename.png')
```
